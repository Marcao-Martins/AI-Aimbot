```python
# Importação das bibliotecas
import torch
import mss
import cv2
import pyautogui
from ultralytics import YOLO
import numpy as np
import os
import csv
import requests
```

```python
# Importa as classes necessárias
from YOLO import YOLODetection

# Inicializa a captura de tela para uma janela específica
# Substitua 'Nome do Jogo' pelo título da janela do jogo

# Caminho para o modelo treinado
trained_model_path = r'C:\Users\Marcos\Desktop\Projetos\AI aimbot\training\runs\train\training_run\weights\best.pt'

# Inicializar a detecção YOLO com o modelo treinado
yolo_detector = YOLODetection(weights_path=trained_model_path, conf_threshold=0.4, valid_classes=['Demonio', 'Zumbi'])
```

```python
# Carrega o vídeo
video_path = r"C:\Users\Marcos\Desktop\Projetos\AI aimbot\output.avi" 
cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    # Lê o frame
    ret, frame = cap.read()
    
    if not ret:
        print("Fim do vídeo")
        break
        
    # Realiza a detecção
    detections = yolo_detector.detect_objects(frame)
    
    # Desenha as detecções no frame
    frame_with_detections = yolo_detector.draw_detections(frame, detections)
    
    # Mostra o frame
    cv2.imshow('Detecções', frame_with_detections)
    
    # Pressione 'q' para sair
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Libera os recursos
cap.release()
cv2.destroyAllWindows()
```
